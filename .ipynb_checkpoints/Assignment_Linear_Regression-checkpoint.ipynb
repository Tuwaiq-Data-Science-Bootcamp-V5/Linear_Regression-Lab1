## Understanding the Data

### `FuelConsumption.csv`:

We have downloaded a fuel consumption dataset, **`FuelConsumption.csv`**, which contains model-specific fuel consumption ratings and estimated carbon dioxide emissions for new light-duty vehicles for retail sale in Canada. [Dataset source](http://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkML0101ENSkillsNetwork20718538-2022-01-01)

*   **MODELYEAR** e.g. 2014
*   **MAKE** e.g. Acura
*   **MODEL** e.g. ILX
*   **VEHICLE CLASS** e.g. SUV
*   **ENGINE SIZE** e.g. 4.7
*   **CYLINDERS** e.g 6
*   **TRANSMISSION** e.g. A6
*   **FUEL CONSUMPTION in CITY(L/100 km)** e.g. 9.9
*   **FUEL CONSUMPTION in HWY (L/100 km)** e.g. 8.9
*   **FUEL CONSUMPTION COMB (L/100 km)** e.g. 9.2
*   **CO2 EMISSIONS (g/km)** e.g. 182   --> low --> 0

The **CO2 EMISSIONS (g/km)** column is the dependent column, and the rest are the independent columns.

# Exercise 4 : Remove Outliers

Then plot a box plot to check if the outliers was removed or not.

```python
Q1 = df.FUELCONSUMPTION_COMB_MPG.quantile(0.25)
Q3 = df.FUELCONSUMPTION_COMB_MPG.quantile(0.75)
IQR = Q3-Q1
upper = Q3 + 1.5*IQR
df.drop(df[(df["FUELCONSUMPTION_COMB_MPG"] > upper)].index, inplace=True)
sns.boxplot(x = df["FUELCONSUMPTION_COMB_MPG"]);
```

```python
Q1 = df.FUELCONSUMPTION_HWY.quantile(0.25)
Q3 = df.FUELCONSUMPTION_HWY.quantile(0.75)
IQR = Q3-Q1
upper = Q3 + 1.5*IQR
df.drop(df[(df["FUELCONSUMPTION_HWY"] > upper)].index, inplace=True)
sns.boxplot(x = df["FUELCONSUMPTION_HWY"]);
```

```python
Q1 = df.FUELCONSUMPTION_CITY.quantile(0.25)
Q3 = df.FUELCONSUMPTION_CITY.quantile(0.75)
IQR = Q3-Q1
upper = Q3 + 1.5*IQR
df.drop(df[(df["FUELCONSUMPTION_CITY"] > upper)].index, inplace=True)
sns.boxplot(x = df["FUELCONSUMPTION_CITY"]);
```

```python
Q1 = df.FUELCONSUMPTION_COMB.quantile(0.25)
Q3 = df.FUELCONSUMPTION_COMB.quantile(0.75)
IQR = Q3-Q1
upper = Q3 + 1.5*IQR
df.drop(df[(df["FUELCONSUMPTION_COMB"] > upper)].index, inplace=True)
sns.boxplot(x = df["FUELCONSUMPTION_COMB"]);
```

```python
Q1 = df.ENGINESIZE.quantile(0.25)
Q3 = df.ENGINESIZE.quantile(0.75)
IQR = Q3-Q1
upper = Q3 + 1.5*IQR
df.drop(df[(df["ENGINESIZE"] > upper)].index, inplace=True)
sns.boxplot(x = df["ENGINESIZE"]);
```
